```python
df1 = pd.DataFrame({'RM' : df[5], 'LSTAT': df[12], 'Price' : df['Price']})
```

```python
data_1 = df1.loc[:,['RM','LSTAT']]
target_1 = df1.loc[:,'Price']
```

```python
from sklearn.preprocessing import StandardScaler
standardized_data_1 = StandardScaler().fit_transform(data_1)
print(standardized_data_1.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(standardized_data_1, 
                                                             target_1, 
                                                             test_size = 0.2,
                                                             random_state = 0)
```

```python
plt.scatter(y_train_1, residual_1)
```

```python
y_test_pred_1 = model_1.predict(X_test_1)
```

```python
temp_df = pd.DataFrame({'Actual': y_test_1, 'Predicted': y_test_pred_1})
temp_df
```

```python
from sklearn import metrics

print('Mean Absolute Error :',metrics.mean_absolute_error(y_test_1, y_test_pred_1))
print('Mean Squared Error :',metrics.mean_squared_error(y_test_1, y_test_pred_1))
print('Root Mean Squared Error :',np.sqrt(metrics.mean_squared_error(y_test_1, y_test_pred_1)))
```

```python
print(metrics.r2_score(y_test_1, y_test_pred_1))
```

###  3rd Polynomial Degree Transformation


```python
# Function for Polynomial Transformation & Fitting data in Linear Regression

def Polynomial_transformation(X, degree):
    # Polynomial Feature Transformation
    poly = PolynomialFeatures(degree)
    poly_data = poly.fit_transform(X)
    return poly_data

def Model_fitting(Poly_data, y):
    # Line Fitting into Data
    model = LinearRegression()
    return model.fit(Poly_data, y)

def get_metrics(y, y_pred):
    from sklearn import metrics
    print('Mean Absolute Error :', metrics.mean_absolute_error(y, y_pred))
    print('Mean Squared Error :', metrics.mean_squared_error(y, y_pred))
    print('Root Mean Squared Error :', np.sqrt(metrics.mean_squared_error(y, y_pred)))
    print('R2 score :', metrics.r2_score(y, y_pred)) 
```

```python
poly_data_3 = Polynomial_transformation(X_train_1, 3)
```

```python
model_3 = Model_fitting(poly_data_3, y_train_1)
print(model_3)
```

```python
# Residual Analyisis of Model_3
y_train_pred_3 = model_3.predict(poly_data_3)
residual_3 = y_train_1 - y_train_pred_3
sns.displot(x = residual_3, kde = True)
```

```python
plt.scatter(y_train_1, residual_3)
```

```python
y_test_pred_3 = model_3.predict(Polynomial_transformation(X_test_1, 3),)
```

```python
get_metrics(y_test_1, y_test_pred_3)
```

```python
residual_test_3 = y_test_1 - y_test_pred_3
```
