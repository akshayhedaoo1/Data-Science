The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. Q1. We have to apply Hypothesis Test on Salary Column where pop_std = 209178.93286 and pop_mean = 308266.11990 on sample of 40 values, whether the mean = 308266 or not

Step 1 : 
         Alternate Hypothesis (Bold Claim) :           $$H_1 : \mu \neq 308266 $$
         Null Hypothesis (Status Quo )     :           $$H_0 : \mu = 308266 $$
                                                       
         
Step 2 : 
        * Collect the Sample size of n = 40 :       $$sample$$
        * Calculate sample mean :                   $$ \bar{x} = ?$$
        
Step 3 :
        Compute Test Statistics :  
                                         $$ t = \frac{\bar{x} - \mu }{\mu/sqrt(n)}\\ $$           
Step 4 :
        Decide $\alpha$
        
Step 5.1 :
        Tailed T test
                                          $$ reject H_0 if ? $$
Step 5.2 :
        Compute p-value :

```python
# Describing the population and sample values
sample_size = 40
sample_mean = 401750.0
pop_mean = 308266
pop_std = 209178.93286
```

```python
# Calclating the z-critical value
confidence_level = 0.90
alpha = 1 - confidence_level
z_critical = norm.ppf(1-alpha/2)
print(z_critical)
```

```python
# Compute the Z score
z_score = z_score(40, 401750, 308266, 209178.9328)
print(z_score)
```

```python
# Plotting the graph values

# Minimum and Maximum X values
x_min = 35000
x_max = 1000000

# Describing the pop_mean and std
mean = pop_mean
std = pop_std/(sample_size ** 0.5)

# Plot the values
x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)
plt.xlim(x_min, x_max)
plt.plot(x, y )

# Calculating the z_critical_left and z_critical_right
z_critical_left = pop_mean - (z_critical * std)
z_critical_right = pop_mean + (z_critical * std)

# Shading the left Rejection region
x1 = np.linspace(x_min, z_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color = 'green')

# Shading the right Rejection region
x2 = np.linspace(z_critical_right, x_max, 100)
y2 = norm.pdf(x2, mean, std)
plt.fill_between(x2, y2, color = 'green')

# Ploting the sample mean and concluding the results 
plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.00000007))
```

```python
# Conclusion using Z test
if (z_score > z_critical):
    print('Reject Null Hypothesis')
else:
    print('Fail to Reject Null Hypothesis')
```

```python
# Conclusion using p test
p_value = 2 * (1 - norm.cdf(np.abs(z_score)))
print(p_value)

if (p_value < alpha):
    print('Reject Null Hypothesis')
else:
    print('Fail to Reject Null Hypothesis')
```

Q2. A person giving a claim of 12th students having the average percentage greater than 65 %. Check whether he is correct or not using T test

Step 1 : 
         Alternate Hypothesis (Bold Claim) :           $$H_1 : \mu > 65 $$
         Null Hypothesis (Status Quo )     :           $$H_0 : \mu <= 65 $$
                                                       
         
Step 2 : 
        * Collect the Sample size of n = 25 :           $$sample$$
        * Calculate sample mean :                       $$ \bar{x} = ?$$
        
Step 3 :
        Compute Test Statistics :  
                                         $$ t = \frac{\bar{x} - \mu }{S/sqrt(n)}\\ $$           
Step 4 :
        Decide $\alpha$
        
Step 5.1 :
        Tailed T test
                                          $$ reject H_0 if ? $$
Step 5.2 :
        Compute p-value :

```python
#Take sample of 12Percentage Column 
sample_value = []
values = random.sample(range(0,df.shape[0]), 25)
sample_value = df.iloc[values,12]
```

```python
sample_value.mean()
```

```python
np.std(df['12percentage'])
```

```python
# Describing the sample and population values
sample_size = 25
sample_mean = 74.668
pop_mean = 65
pop_std = 11.05523
```

```python
# Compute the t-critical value
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = t.ppf(1 - alpha, df = 24)
print(t_critical)
```

```python
# Compute the t_score value
t_score = z_score(25, 74.668, 65, 11.055)
print(t_score)
```

```python
# Plotting the graph values

# Minimum and Maximum X values
x_min = 50
x_max = 90

# Describing the pop_mean and std
mean = pop_mean
std = pop_std/(sample_size ** 0.5)

# Plot the values
x = np.linspace(x_min, x_max, 100)
y = t.pdf(x, loc = mean, scale = std, df= 24)
plt.xlim(x_min, x_max)
plt.plot(x, y )

# Calculating the z_critical_right
z_critical_right = pop_mean + (z_critical * std)


# Shading the right Rejection region
x1 = np.linspace(z_critical_right, x_max, 100)
y1 = t.pdf(x1, loc = mean, scale = std, df = 24)
plt.fill_between(x1, y1, color = 'red')

# Ploting the sample mean and concluding the results 
plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.00000007))
```

```python
# Conclusion using Z test
if (t_score > t_critical):
    print('Reject Null Hypothesis')
else:
    print('Fail to Reject Null Hypothesis')
```

```python
# Conclusion using P test
p_value = 1 - t.cdf(np.abs(t_score), df = 25)
print(p_value)

if (p_value < alpha):
    print('Reject Null Hypothesis')
else:
    print('Failed to Reject Null Hypothesis')
```
