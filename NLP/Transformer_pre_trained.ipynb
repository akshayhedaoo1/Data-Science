# Zero shot Classification - NER-CLASSIFICATION

```python
from transformers import pipeline
generator = pipeline('zero-shot-classification')
generator("This course is about Deep learning at Innomatics", candidate_labels = ['education', 'politics', 'business'])
```

```python
import wikipedia
from transformers import pipeline
covid = wikipedia.page('Covid pandemic',)
generator = pipeline('text-generation')
generator(covid.section('Epidemiology'))
```

```python
from transformers import pipeline
generator = pipeline('text-generation', model = "distilgpt2")
generator(covid.section('Epidemiology'),max_length = 10, num_return_sequences=2)
```

# Question Answering 

```python
from transformers import pipeline
question_answer = pipeline('question-answering')
question_answer(question="Where do i got training on DL", context="My name is Rishav, i learned DL from Innomatics")
```

# Fill in the blanks (MASKING)

```python
from transformers import pipeline
FILL = pipeline('fill-mask')
FILL("I want to learn <mask> learning from Innomatics", top_k=2)
```

# NER

```python
from transformers import pipeline
NER = pipeline('ner', grouped_entities = True)
NER("I am studying NLP from Innomatics ")
```

# Text Summarization

```python
import wikipedia
from transformers import pipeline
covid = wikipedia.page('Covid pandemic')
summarization = pipeline('summarization')
summarization(covid.section('Epidemiology'))
```
